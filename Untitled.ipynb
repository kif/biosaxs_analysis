```python
%matplotlib widget
import json
import numpy
import hdf5plugin
import h5py
import pyFAI
from pyFAI.gui import jupyter
from matplotlib.pyplot import subplots
```

```python
keep = slice(0,9)
with h5py.File("sample_BSA_5mg0000-integrate.h5","r") as h:
    diode = h["/entry_0000/0_measurement/diode"][keep]
    frames = h["/entry_0000/0_measurement/images"][keep]
    config = json.loads(h["/entry_0000/1_integration/configuration/data"][()])
print(diode)
jupyter.display(frames[2])
```

```python
ai = pyFAI.load(config)
ai
```

```python
ai.detector.mask = numpy.load("mask_17Feb22a.npy")
```

```python
kwargs = {"npt": 1000, "method": ("no", "csr", "cython"), "error_model":"poisson"}

# size = 10
diode = numpy.random.randint(1, 10, size=len(frames))
# frames = [frames[0]*d for d in diode]
```

```python
#Integrate first, then average curves
int_first = [ ai.integrate1d(frame, normalization_factor=norm, **kwargs) 
             for frame, norm in zip(frames, diode)]
q1 =int_first[0].radial
I1 = numpy.mean([i.intensity for i in int_first], axis=0)
sigma1 = numpy.sqrt(numpy.sum([i.sigma**2 for i in int_first], axis=0))/len(int_first)
```

```python
fig,ax = subplots()
ax.errorbar(q1, I1, sigma1, label="integrate then average")
ax.set_yscale("log")
ax.set_xlabel(int_first[0].unit.label)
ax.set_ylabel("Intensity")
```

```python
#Average first, then integrate
img_sum = numpy.sum(frames, axis=0)
nrm_sum = sum(diode)
q2, I2, sigma2 = ai.integrate1d(img_sum, normalization_factor=nrm_sum, **kwargs)
```

```python
fig,ax = subplots()
ax.errorbar(q2, I2, sigma2, label="average then integrate")
ax.set_yscale("log")
ax.set_xlabel(int_first[0].unit.label)
ax.set_ylabel("Intensity")
```

```python
fig,ax = subplots()
ax.plot(q1, I1-I2)
```

```python
fig,ax = subplots()
ax.plot(q1, sigma1-sigma2, label="delta")
ax.plot(q1, sigma1 , label="integrate then average")
ax.plot(q2, sigma2, label="average then integrate")
ax.legend()
```
